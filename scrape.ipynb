```python
#final code

#import packages
import numpy as np
import pandas as pd
import requests 

#naming object
source = requests.get("https://www.bovada.lv/services/sports/event/v2/events/A/description/football/nfl").json()
data = source[0]

#lots of code to find the proper data elements

#iterate through
selected_elements = []

#using range to skip over error row
for i in data['events']:
  row = []
  row.append(i['competitors'][0]['name'])
  row.append(i['competitors'][1]['name'])
  row.append(float(i['displayGroups'][0]['markets'][1]['outcomes'][1]['price']['handicap']))
  row.append(float(i['displayGroups'][0]['markets'][2]['outcomes'][0]['price']['handicap']))
  selected_elements.append(row)

df = pd.DataFrame(selected_elements, columns = ['Home','Away','spread','o/u'])
df.head(10)
```
